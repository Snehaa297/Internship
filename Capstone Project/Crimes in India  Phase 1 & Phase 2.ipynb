```python
state_crime = pd.read_csv('/Users/sneha/Downloads/State_Crime.csv')
state_crime
```

```python
crime_2011_df = state_crime[state_crime['YEAR'] == 2011]
crime_2011_df
```

```python
# Select columns for analysis (from the 2nd column to the third last column representing crime types)
crime_columns = crime_2011_df.columns[2:-4]
crime_columns
```

```python
literacy_rate_column =crime_2011_df['LiTERACY RATE ']
literacy_rate_column
```

```python
# Finding the column with the maximum value for each row
max_crime_column = crime_2011_df[crime_columns].idxmax(axis=1)

# Adding a new column to the DataFrame to store the column name with the maximum value
crime_2011_df['max_crime_column'] = max_crime_column

# Displaying the updated DataFrame
```

```python
crime_2011_df
```

```python
# Finding the top 2 columns with the highest values for each row
top_5_crime_columns = crime_2011_df[crime_columns].apply(lambda row: row.nlargest(5).index.tolist(), axis=1)

# Adding new columns to the DataFrame to store the top 2 column names with the highest values
crime_2011_df[['top_crime_column_1', 'top_crime_column_2','top_crime_column_3','top_crime_column_4','top_crime_column_5']] = pd.DataFrame(top_5_crime_columns.tolist(), index=crime_2011_df.index)
```

```python
crime_2011_df
```

```python
for index, row in crime_2011_df.iterrows():
    top_5_crime_columns = [
        row['top_crime_column_1'],
        row['top_crime_column_2'],
        row['top_crime_column_3'],
        row['top_crime_column_4'],
        row['top_crime_column_5']
    ]

    values = [row[top_5_crime_columns[0]], row[top_5_crime_columns[1]],
              row[top_5_crime_columns[2]], row[top_5_crime_columns[3]],
              row[top_5_crime_columns[4]]]

    colors  = sns.color_palette("Blues_d", n_colors=5)

    plt.bar(top_5_crime_columns, values, color = colors)
    plt.xlabel('Crime Columns')
    plt.ylabel('Number of Cases')
    plt.title(f'Top 5 Crime Columns - {row["STATE/UT"]}')
    plt.xticks(rotation = 90)
    plt.show()
```

```python

# Select only the crime type columns
crime_columns = crime_2011_df.columns[2:-6]

# Set a color palette with different colors for each state
state_colors = sns.color_palette('husl', n_colors=len(crime_2011_df))

# Create a scatter plot for each crime type column
for crime_column in crime_columns:
    plt.figure(figsize=(8, 6))  # Adjust the figure size as needed
    
    for index, row in crime_2011_df.iterrows():
        # Plot scatter for each state with different colors
        plt.scatter(row['LiTERACY RATE '], row[crime_column], label=row['STATE/UT'])
    plt.xlabel('Literacy Rate')
    plt.ylabel(f'Number of Cases - {crime_column}')
    plt.title(f'Scatter Plot: Literacy Rate vs {crime_column}')
    #plt.legend(title='State', bbox_to_anchor=(1,0.5),loc='center left')
    plt.grid(True)
    plt.show()
```
