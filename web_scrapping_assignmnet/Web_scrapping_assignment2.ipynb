# 1. Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.
This task will be done in following steps:
1. First get the webpage https://www.naukri.com/
2. Enter “Data Analyst” in “Skill, Designations, Companies” field and enter “Bangalore” in “enter the
location” field.
3. Then click the searchbutton.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.

```python
#  First get the webpage https://www.naukri.com/
driver = webdriver.Chrome(r'chromedriver.exe')
```

```python
driver.get('https://www.naukri.com')
```

```python
# Enter “Data Analyst” in “Skill, Designations, Companies” field and enter “Bangalore” in “enter the location” field

designation =driver.find_element(By.CLASS_NAME,'suggestor-input')
designation.send_keys("Data Analyst")
```

```python
# Enter "Bangalore" in location field.
location = driver.find_element(By.XPATH,'/html/body/div[1]/div[7]/div/div/div[5]/div/div/div/div[1]/div/input')
location.send_keys("Bangalore")
```

```python
# Then click the searchbutton.
search = driver.find_element(By.CLASS_NAME,'qsbSubmit')
search.click()
```

```python
# Then scrape the data for the first 10 jobs results you get.

### To scrape data separately, creating an empty list for each feature in dataframe. 
job_title=[]
company_name=[]
expereience=[]
salary=[]
job_location=[]
job_description =[]
job_desc_key =[]
job_posted_date =[]
job_url=[]

tags = driver.find_elements(By.XPATH,'//a[@class="title ellipsis"]')
for i in tags[0:10]:
    tag = i.text
    job_title.append(tag)

    
tags = driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')
for i in tags[0:10]:
    tag = i.text.strip('\n')
    company_name.append(tag)


tags = driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft expwdth"]')
for i in tags[0:10]:
    tag = i.text
    expereience.append(tag)

tags = driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft "]')
for i in tags[0:10]:
    tag = i.text
    salary.append(tag)

tags = driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft locWdth"]')
for i in tags[0:10]:
    tag = i.text
    job_location.append(tag)

tags = driver.find_elements(By.XPATH,'//div[@class="ellipsis job-description"]')
for i in tags[0:10]:
    tag = i.text
    job_description.append(tag)

tags = driver.find_elements(By.XPATH,'//ul[@class="tags has-description"]')
for i in tags[0:10]:
    tag = i.text
    job_desc_key.append(tag)

tags = driver.find_elements(By.XPATH,'//span[@class="fleft postedDate"]')
for i in tags[0:10]:
    tag = i.text
    job_posted_date.append(tag)
    

url = driver.find_elements(By.XPATH,'//a[@class="title ellipsis"]')
for i in url[0:10]:
    tag = i.get_attribute('href')
    job_url.append(tag)

```

```python
### creating a dataframe and putting the lists of scrapped data  in the dataframe

data_analyst_10 =pd.DataFrame(list(zip(job_title,company_name,expereience,salary,job_location,job_description,job_desc_key,job_posted_date,job_url)),
                              columns=['Job_title','Company','Exp.','Salary','Location','Description','Skills','Postings date','URL'])

## for job_url making the column clickable.

def make_clickable(val):
    # target _blank to open new window
    return '<a target="_blank" href="{}">{}</a>'.format(val, val)

## display dataframe
data_analyst_10.style.format({'URL': make_clickable})
```

```python
driver.close()
```

```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Set up Selenium webdriver
driver = webdriver.Chrome()

# Open Flipkart.com
driver.get("https://www.flipkart.com/")

# Close login popup if present
try:
    driver.find_element(By.XPATH, "//button[text()='✕']").click()
except:
    pass

# Search for sunglasses
search_query = "sunglasses"
search_box = driver.find_element(By.NAME, "q")
search_box.send_keys(search_query)
search_box.submit()

# Wait for the search results to load
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "_1AtVbE")))

# Scrape data
data = []
while len(data) < 100:
    # Wait for the products to load
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "_2kHMtA")))

    # Get product details
    product_names = driver.find_elements(By.CLASS_NAME, "_4rR01T")
    prices = driver.find_elements(By.CLASS_NAME, "_30jeq3")
    descriptions = driver.find_elements(By.CLASS_NAME, "_1I_Atq")

    # Store product details in a list
    for i in range(len(product_names)):
        product_name = product_names[i].text
        price = prices[i].text
        description = descriptions[i].text
        data.append([product_name, price, description])

        if len(data) == 100:
            break

    # Click on the "Next" button
    try:
        next_button = driver.find_element(By.XPATH, "//a[text()='Next']")
        driver.execute_script("arguments[0].scrollIntoView();", next_button)
        next_button.click()
    except:
        break

    time.sleep(2)

# Create a dataframe from the scraped data
df = pd.DataFrame(data, columns=["Product Name", "Price", "Description"])

# Print the dataframe
print(df)

# Close the browser
driver.quit()
```

This task will be done in following steps:
1. First get the webpagehttps://www.jagranjosh.com/
2. Then You have to click on the GK option
3. Then click on the List of all Prime Ministers of India
4. Then scrap the mentioned data and make theDataFrame.

```python
driver = webdriver.Chrome(r'chromedriver.exe')
```

```python
driver.get('https://www.jagranjosh.com/')
```

```python
driver.find_element(By.LINK_TEXT,'GK').click()
```

```python
driver.find_element(By.LINK_TEXT,'List of all Prime Ministers of India').click()
```

```python
# Find the table containing the Prime Ministers' list
table = driver.find_element(By.TAG_NAME,"tbody")
# Find all rows in the table
rows = table.find_elements(By.TAG_NAME,"tr")
```

```python
data = []

# Extract the data from each row and append it to the DataFrame
for row in rows[1:]:
    columns = row.find_elements(By.TAG_NAME,"td")
    prime_minister = columns[1].text
    born_dead = columns[2].text
    Term_of_Office = columns[3].text.replace('\n','|')
    Remark =columns[4].text
    data.append([prime_minister, born_dead, Term_of_Office, Remark])

df = pd.DataFrame(data, columns=["Prime Minister", "Born-Dead", "Term Of Office","Remark"])
```
